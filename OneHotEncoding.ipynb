```python
# Build a one-hot encoding function using tensors.
import torch
```

```python
classes = torch.randint(0, 10, (1, 5))
classes = classes.squeeze()
```

```python
def one_hot_encoding(classes, numclasses = None):
  if numclasses is None:
        numclasses = max(classes) + 1

  ohe = torch.zeros(len(classes), numclasses)
  for i, cls in enumerate(classes):
    ohe[i][cls] = 1
  return ohe
ohe = one_hot_encoding(classes)
print(ohe)
```
