```python
# Write a function to normalize tensors (min-max scaling).
```

```python
import torch
```

```python
mat = torch.randint(0,10, (3,3))
mat
```

```python
def noramlize(t):
  max = torch.max(t)
  min = torch.min(t)
  updated_t = (t - min)/(max - min)
  return updated_t

noramlize(mat)
```

```python
# column wise normalization
def normalize_by_columns(t):
  max = t.max(dim = 1).values
  min = t.min(dim = 1).values
  updated_t = (t - min)/(max - min )
  return updated_t

normalize_by_columns(mat)
```
