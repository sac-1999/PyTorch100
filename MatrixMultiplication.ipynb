```python
# 1. Fundamentals & Tensor Operations (10) Implement matrix multiplication manually using PyTorch tensors (no torch.mm).
```

```python
import torch
```

```python
r1 = 3
c1 = 4
c2 = 1
mat1 = torch.randint(0,10,(r1,c1))
mat2 = torch.randint(0,10,(c1,c2))
finalmat = torch.rand((r1,c2))
```

```python
for i, row in enumerate(mat1):
  for j in range(c2):
    t1 = row
    col = mat2[:, j]
    finalmat[i][j] = torch.dot(t1, col)

```
