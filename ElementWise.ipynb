```python
# Implement element-wise operations without built-in functions.
```

```python
import torch
```

```python
mat1 = torch.randint(0,10, (1,3))
mat2 = torch.randint(0, 10, (3, 3))
```

```python
mat1 * mat2
```

```python
mat1, mat2
```

```python
def element_wise_multiplication(mat1, mat2):
  assert mat1.shape == mat2.shape
  mat3 = torch.randint(0, 10, mat1.shape)
  for i in range(mat1.shape[0]):
    for j in range(mat1.shape[1]):
      mat3[i][j] = mat1[i][j] * mat2[i][j]

  return mat3

element_wise_multiplication(mat1, mat2)
```

```python
torch.mul(mat1, mat2)
```
