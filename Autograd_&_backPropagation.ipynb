```python
# Custom Autograd & Backpropagation (10) Build a custom autograd function for ReLU.
```

```python
import torch
```

```python
class square_f(torch.autograd.Function):
  @staticmethod
  def forward(ctx, x):
    ctx.save_for_backward(x)
    return x**2
  @staticmethod
  def backward(ctx, grad_output):
    print('--------', ctx.saved_tensors)
    print(grad_output)
    (x,) = ctx.saved_tensors
    return 2*x * grad_output
```

```python
class cube_f(torch.autograd.Function):
  @staticmethod
  def forward(cxt, x):
    cxt.save_for_backward(x)
    return x**3

  @staticmethod
  def backward(cxt, grad_output):
    (x, ) = cxt.saved_tensors
    return 3*x*grad_output
```

```python
x = torch.arange(1,10, 1,dtype=torch.float32, requires_grad = True)
y = square_f.apply(x)
y2 = cube_f.apply(y)
```

```python
y, y2, x
```

```python
y2.backward(torch.ones(1,9).squeeze())
```
