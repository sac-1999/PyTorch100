```python
# Write a script to compute cosine similarity between two tensors.
import torch
```

```python
def cosine_similarity(t1, t2):
  dotproduct = torch.dot(t1, t2)
  norm1 = torch.norm(t1)
  norm2 = torch.norm(t2)
  return dotproduct/(norm1 * norm2)

t1 = torch.randn(5)
t2 = torch.randn(5)
cosine_similarity(t1, t2)
```
