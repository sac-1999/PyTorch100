```python
# Build a custom autograd function for ReLU.
```

```python
import torch
```

```python
class relu(torch.autograd.Function):
  @staticmethod
  def forward(ctx, x):
    ctx.save_for_backward(x)
    return torch.clamp(x, 0)

  @staticmethod
  def backward(ctx, grad_output):
    (x,) = ctx.saved_tensors
    grad_x = grad_output.clone()
    grad_x[x <= 0] = 0
    return grad_x
```

```python
x = torch.arange(-10,10, 3, dtype = torch.float32, requires_grad = True)
```

```python
y = relu.apply(x)
```

```python
y.backward(torch.ones(x.shape).squeeze())
```
