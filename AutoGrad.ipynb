```python
import torch
```

```python
t1 = torch.tensor(1.0, requires_grad = True)
t2 = torch.tensor(2.0, requires_grad = True)
```

```python
y = t1**3 + t2
```

```python
y.backward()
```

```python
t1.grad, t2.grad
```

```python
t3 = torch.tensor([1.0,2.0,3.0,4.0], requires_grad=True)
```

```python
y1 = t3 ** 2
```

```python
y1.backward(torch.ones(*t3.shape))
```

```python
t3.grad
```
